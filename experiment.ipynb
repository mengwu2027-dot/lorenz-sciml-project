## 1. The Lorenz System
The Lorenz system is defined by the following ODEs:
$$\frac{dx}{dt} = \sigma(y - x)$$
$$\frac{dy}{dt} = x(\rho - z) - y$$
$$\frac{dz}{dt} = xy - \beta z$$
We use standard chaotic parameters: $\sigma = 10, \rho = 28, \beta = 8/3$. Ground truth data is generated via the RK45 solver (see `src/` directory for full implementation). Below is the ground truth trajectory generated via the RK45 numerical solver (2000 steps), which serves as the target manifold for all our SciML models.

## 2. Baseline Model: Multi-Layer Perceptron (MLP)
The MLP is trained to learn the discrete mapping $x_{t+1} = f(x_t)$. Because it lacks continuous-time inductive biases, small numerical errors rapidly accumulate, leading to catastrophic divergence.

![MLP Prediction](figures/mlp_prediction.png)

## 3. Physics-Informed Neural Network (PINN)
The PINN minimizes both data loss and physics residuals. For example, the x-component residual is: $\mathcal{L}_{res,x} = \| \frac{d\hat{x}}{dt} - \sigma(\hat{y} - \hat{x}) \|^2$. While it interpolates perfectly, it fails to extrapolate due to extreme chaotic sensitivity.

![PINN Prediction](figures/pinn_prediction.png)

## 4. Neural Ordinary Differential Equations (Neural ODEs)
Neural ODEs parameterize the continuous derivative: $\frac{d\mathbf{h}(t)}{dt} = f_\theta(\mathbf{h}(t), t)$. This naturally constrains the trajectory to the physical manifold. However, under MSE loss, it collapses into a stable limit cycle to avoid phase-shift penalties.

![Neural ODE Prediction](figures/neural_ode_prediction.png)

## 5. Conclusion
1. **Data alone is insufficient:** Standard deep learning (MLP) diverges wildly.
2. **Physics residuals have limits:** PINNs interpolate beautifully but fail to extrapolate chaos.
3. **Continuous integration stabilizes manifolds:** Neural ODEs respect the topological boundaries of the attractor but require advanced loss functions to capture aperiodic chaotic transitions.
